```python
import pandas as pd
import seaborn as sns 
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
auto = pd.read_csv('Automobile+%281%29.csv')
```

```python
auto.head()
```

```python
#Univariate distribution
sns.displot(auto['city_mpg'], kde=True, rug=True)
plt.show()
```

```python
#Univariate distribution
sns.displot(auto['normalized_losses'], kde=False, rug=True)
plt.show()
```

```python
#Bi-variate dstribution using a residual plot (kind='reg')
sns.jointplot(auto['engine_size'], auto['normalized_losses'])
plt.show()
```

```python
#Bi-variate dstribution using a residual plot (kind='reg')
sns.jointplot(auto['engine_size'], auto['normalized_losses'], kind="reg")
plt.show()
```

```python
#creating a maxtrix of visualization to explain relationship between variables - MULTI-VARIATE distribution
sns.pairplot(auto[['normalized_losses', 'engine_size', 'horsepower']])
```

```python
#plotting a 2d plot(i.e where one plot is a continous variable{int} and the other is categorical{str}) - use the stripplot
sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True)
```

```python
sns.swarmplot(auto['fuel_type'], auto['horsepower'])
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

```python
sns.barplot(auto['body_style'], auto['horsepower'], hue=auto['engine_location'])
```

```python
sns.countplot(auto['body_style'], hue=auto['engine_location'])
```

```python
sns.pointplot(auto['fuel_system'], auto['horsepower'], hue=auto['number_of_doors'])
```

```python
sns.factorplot(x='fuel_type', 
               y='horsepower', 
               hue='number_of_doors', 
               col='engine_location', 
               data= auto,
               kind='swarm')
```

```python
#Linear regression to check the linear relationship between 2 variables
sns.lmplot(x='horsepower', y='peak_rpm', data=auto)
```

```python
sns.lmplot(x='horsepower', y='peak_rpm', data=auto, hue='fuel_type')
```
